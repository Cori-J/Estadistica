# 1. Conceptos basicos. <a name="id1"></a>
- **Crédito**: es todo activo de riesgo, cualquiera sea la modalidad de su instrumentación, mediante el cual la entidad, asumiendo el riesgo de su recuperación, provee o se compromete a proveer fondos u otros bienes o garantiza frente a terceros, el cumplimiento de obligaciones contraídas por su cliente.
- **Riesgo de credito**: probabilidad de que un deudor incumpla, en cualquier grado, con el repago de su(s) obligación(es).
- **Gestion de riesgo de credito**: proceso de identificación, medición, monitoreo, control, y divulgación del riesgo de crédito.
- **Incumplimiento**: se entiende por incumplimiento del prestatario al no pago de los montos adeudados de capital e intereses dentro del período predeterminado o de acuerdo con lo establecido en el plan de pagos pactado
- **Perdida por riesgo de credito**: esta se divide en perdidas esperadas e inesperadas.
- **Perdida esperada**: se la define como la media de la distribucion estadistca de perdidas.
- **Perdida inesperada**: la media de dispercion de la distribucion estadistica de perdidas (variaza, desviacion estandar o percentil considerando un cierto nivel de confianza)  
- **Scoring numerico**: es una medida de tipo ordinal que compara distintas operaciones dentro de una cartera o portafolio.
- **Probabilidad de incumplimiento**: medida estandarizada que, siendo una probabilidad, se encuentra entre 0 % y 100 %.

- **Criterios minimos para la evaluacion de riesgo**:
 
  - Factores generales:
    
    - Indicadores macroeconomicos
    - Analisis sectorial
    - Analisis grupal

  - Factores individuales:

    - Evaluacion socieconomica (capacidad de pago)
    - Comportamiento de pago (buros de informacion crediticia)
- **Definicion por tipo de credito**:
    
    - Creditos individuales
    - Creditos masivos

# 2. Metodos para identificar el riesgo. <a name="id2"></a>

El metodo mas ampliamente utilizado es el de **puntajes tecnicos de scoring**, este metodo consiste basicamente en hacer predicciones de comportamiento con base a los datos historicos de cumplimiento e incumplimiento, estas predicciones se pueden interpretar como probabilidades o puntaje (numerico) scoring.

En lo general existen dos tipos **scoring**:
  - De aprobacion o evaluacion de solicitudes crediticias
  - De gestion de comportamiento, tiene el objetivo de realizar seguimiento a los clientes ya incorporados, tales como:
    - Estimacion de previsiones.
    - Acciones de cobranza.
    - Deteccion preventiva de fraude.
    - Analisis de mercadeo.
    - Proveer ratings de diferentes clientes.

# 3. Metodos para cuantificar el riesgo. <a name="id3"></a>

Los modelos para desarrollar un scoring son los siguientes:
  - Regresion Lineal.
  - Regresion logistica.
  - Analisis discriminante.
  - Arboles de desiciones.
  - Redes neuronales.
  - Algoritmos geneticos.

En los modelos de scoring se pueden encontrar 3 tipos de variables, cuantitavas, cualitativas y cruzadas.

Se debe entrenar un modelo de scoring para cada segmento de la cartera, es decir:
  - Modelo de scoring para personas: las variables que pueden ayudar a explicar la variable dependiente pueden ser: sociodemograficas, comportamiento de pago, comportamiento del cliente, etc.
  - Modelo de scoring para empresas: las variables que pueden ayudar a explicar la variable dependiente pueden ser: retorno sobre activos, retorno sobre patrimonio, utilidad, nivel de ventas, tamano de empresa, indice de liquidez, nivel de endeudamiento, etc.

Por otro lado, tambien se pueden desarrollar modelos de scoring por tipo de producto financieros, los productos mas comunes son:

  - Credito de consumo
  - Credito Vehicular
  - Credito de vivienda
  - Microcredito
    - Capital operativo
    - Capital de inversion
  - Pyme
    - Capital operativo
    - Capital de inversion
  - Gran empresa
    - Capital operativo
    - Capital de inversion

Existen dos metodos para cuantificar el riesgo:
  - Metodo estandar: la medición del riesgo de crédito de los activos se realiza con base en calificaciones o ratings externos otorgados por una empresa calificadora de riesgo de reconocido prestigio internacional.
  - Metodo basado en calificaciones internas (IRB): bajo un esquema de calificaciones internas, la EIF puede calcular los componentes del riesgo
de una determinada operación y determinar el requerimiento de capital y previsiones correspondientes a esa exposición.
    - Componentes del riesgo:
      - ***Probabilidad de incumplimiento (PI)***: este se puede derivar como la proporcion de la frecuencia historica de incumplidos para tipo de cartera o producto, sin embargo esta no puede descriminar segun cliente, por lo que se utiliza una de las tecnicas anteriormente mencionada.
      - ***Perdida dado el incumplimiento (PDI)***: este indicador de severidad mide la pérdida que se sufrira después de haber realizado todas las gestiones para recuperar los créditos que han sido incumplidos y puede estimarse como $1 - r$, donde “$r$” es la tasa de recuperación (en porcentaje). Se entiende como tasa de recuperación a la relación entre el saldo incumplido, neto de recuperaciones por concepto de capital y otros, y el saldo total incumplido.
      Tambien se la puede entender como la diferencia entre el monto adeudado y el valor presente de las garantias que respalda la operacion.
      - ***Exposicion al momento del incumplimiento (E)***: es el monto total comprometido por el deudor en el momento en que ocurre el incumplimiento.

Matematicamente se puede expresa lar ***perdida esperada (PE)*** que representa una medida promedio, como:
\begin{equation}
  PE = PI * PDI * E
\end{equation}
Esta es cubierta con prevision.

Por otro lado, la **perdida inesperada (PIN)** es una medida de dispersión respecto a la media de las pérdidas por riesgo de crédito, esta se cubre con capital.

El codigo siguiente basicamente lo que hace es dividir el conjunto de datos en dos (train y test).

Posteriormente se se hace realiza 3 transformaciones a los datos categoricos:
 - Se convierte los datos categoricos a numericos, creando una variable dummy (en machine learning es conocido como One-hot-encoder) para cada subcategoria, por e.j. la variable genero se divide en dos ya que esta tiene dos subclases, masculino y femenino, las cuales son remplazados por 1 si pertenece a masculino y 0 en otro caso.
 - Se convierte los datos categoricos a numericos, pero esta vez sin crear variables dummy, sino se remplaza las subcategorias en la misma columna o categoria (este metodo es conocido en machine learning como Label Econder.
 - La ultima trasnformacion que se realiza es una mezcla de las transformaciones anteriores, solo que esta vez creamos variables dummy para todas las variables categorias excepto para ```genero``` e ```historial_cre``` las cuales son convertidas por el metodo Label encoder.



````python
# Separacion de variables explicativas y cuantitativas
y = data.repago
X = data.drop('repago', axis = 1)

# Division de datos en entrenamiendo y test

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    train_size = .8,
                                                    test_size = .2,
                                                    random_state = 0,
                                                    stratify = y
                                                    )

# Obtencion de variables categoricas con one hot

col_cat_oh = ['genero', 'historial_cre']

# Obtencion de variables categoricas con label encoder

col_cat_le = [col for col in X.columns if X[col].dtype == 'object']
col_cat_le = list(set(col_cat_le) - set(col_cat_oh)) 

# Variables categorias totales

col_cat_tt = col_cat_le + col_cat_oh 

# Obtencion de variables numericas

col_con = [col for col in X.columns if X[col].dtype in ['int64',
                                                        'float64']] 


# preprocesamiento v. categoricas para train y test (todas con dummis)

X_train_catd = pd.get_dummies(X_train[col_cat_tt], drop_first = True)
X_test_catd = pd.get_dummies(X_test[col_cat_tt], drop_first = True)

# Union de v. categoricas y continuas para train y test (todas con dummis)

hot_X_train = pd.concat([X_train_catd, X_train[col_con]], axis = 1)
hot_X_test = pd.concat([X_test_catd, X_test[col_con]], axis = 1)


# preprocesamiento v. categoricas para train y test  (label encoder)

label_encoder = LabelEncoder()

label_X_train = X_train.copy()
label_X_test = X_test.copy()

for col in col_cat_tt:
    label_X_train[col] = label_encoder.fit_transform(X_train[col])
    label_X_test[col] = label_encoder.transform(X_test[col])

# preprocesamiento V_categoricas para train y test (mix)

string = 'estabilidad_lab|otros_deb|activos|cta_cte'
Logic =  hot_X_train.columns.str.contains( string , regex = True)

mix_X_train = pd.concat([
                         label_X_train.drop(col_cat_oh + col_con, axis = 1),
                         hot_X_train[hot_X_train.columns[Logic == False]]
                         ], 
                        axis = 1)

mix_X_test = pd.concat([
                         label_X_test.drop(col_cat_oh + col_con, axis = 1),
                         hot_X_test[hot_X_train.columns[Logic == False]]
                         ], 
                        axis = 1)
````

En el codigo siguiente se encuentra los modelos que utilizaremos para poder estimar las probabilidades de si una persona pagara su credito o no.

````python
# Clase para optimizar codigo

class Score():
  def __init__(self, y_train, X_train, y_test, X_test):

    self.y_train = y_train
    self.X_train = X_train
    self.y_test = y_test
    self.X_test = X_test

  def model_l_s(self):
    """
    Devuelve un modelo logistico entrenado con el modulo Statsmodels
    """
    model = sms.Logit(self.y_train, 
                      sms.add_constant(self.X_train)).fit()
    
    return model

  def pvalues(self, alpha = 0.05):
    '''
    Retorna las variables significativas a un 95% de confianza.
    '''
    model = self.model_l_s()
    pvalues = model.params[model.pvalues < alpha]
    return pvalues

  def model_lg_sk(self):
    """
    Devuelve los valores predecidos de un modelo losgistico
    entrenado con el modulo Sklearn
    """
    # modelo logistico
    lg = LogisticRegression(penalty = 'l2', class_weight = 'balanced',
                          solver = 'lbfgs')
    
    lg.fit(self.X_train, self.y_train)
    prob = lg.predict_proba(self.X_train)
    pred = lg.predict(self.X_test)
    return pred, prob

  def model_svc_sk(self):
    """
    """
    from sklearn.svm import SVC

    svc = SVC()
    svc.fit(self.X_train, self.y_train)
    pred_svc = svc.predict(self.X_test)
    return pred_svc

  def model_nn_sk(self):
    from sklearn.neighbors import KNeighborsClassifier

    knc = KNeighborsClassifier()
    knc.fit(self.X_train, self.y_train) 
    pred_knc = knc.predict(self.X_test)
    return pred_knc

  def model_nb_sk(self):
    from sklearn.naive_bayes import GaussianNB

    clf = GaussianNB()
    clf.fit(self.X_train, self.y_train)
    pred_clf = clf.predict(self.X_test)
    return pred_clf 

  def model_tree_sk(self):
    from sklearn.tree import DecisionTreeClassifier

    dtc = DecisionTreeClassifier()
    dtc.fit(self.X_train, self.y_train)
    pred_dtc = dtc.predict(self.X_test)
    return pred_dtc

  def model_bo(self):
    """
    Devuelve los valores predecidos de un modelo Gradient Tree Boosting
    del modulo sklearn, los parametros estan optimizados.
    """

    param_grid = {'n_estimators'  : [50, 100, 500, 1000],
                'max_features'  : ['auto', 'sqrt', 'log2'],
                'max_depth'     : [None, 1, 3, 5, 10, 20],
                'subsample'     : [0.5, 1],
                'learning_rate' : [0.001, 0.01, 0.1]
              }
    grid = GridSearchCV(
          estimator  = GradientBoostingClassifier(random_state=123),
          param_grid = param_grid,
          scoring    = 'accuracy',
          n_jobs     = multiprocessing.cpu_count() - 1,
          cv         = RepeatedKFold(n_splits=3, n_repeats=1, random_state=123), 
          refit      = True,
          verbose    = 0,
          return_train_score = True
        )
    grid.fit(X = self.X_train, y = self.y_train)
    modelo_final = grid.best_estimator_
    predicciones = modelo_final.predict(X = self.X_test)
    probabilidades = modelo_final.predict_proba(X = self.X_test)
    return predicciones, probabilidades

def metricas (y_test, pred, colname):
  """
  Devuelve diferentes metricas para el caso de clasificacion. 
  """

  # Presicion: ¿Qué proporción de identificaciones positivas fue correcta?
  # Exhaustividad ¿Qué proporción de positivos reales se identificó correctamente?

  pre_s = precision_score(y_test, pred)
  rec_s = recall_score(y_test, pred)
  acu_s = balanced_accuracy_score(y_test, pred)
  f1_s = f1_score(y_test, pred)
  ck_s = cohen_kappa_score(y_test, pred)
  roc_auc = roc_auc_score(y_test, pred)

  indice = ['Precision_score', 'Recall_score', 'b_accuracy_score',
            'f1_score', 'cohen_kappa_score', 'roc_auc_score']
  
  column = [pre_s, rec_s, acu_s, f1_s, ck_s, roc_auc]
  metricas_tab = pd.DataFrame(data = {colname :column}, index = indice)
  return metricas_tab
````

Se puede evidenciar que el metodo One hot enconder nos proporciona mas variables significativas a la hora de correr un modelo logistico.
