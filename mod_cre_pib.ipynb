# 1. Introducción.

El sistema financiero tiene la función de captar los recursos de los agentes superavitarios y canalizarlos hacia los agentes deficitarios, un medio para realizar dicha función es atreves del crédito que tiene el fin de financiar el capital de inversión, capital de trabajo o en consumo. Por otro lado, el crecimiento económico no es más que un aumento en la producción o productividad, esta última no necesariamente, de bienes y servicios en un determinado periodo, 1 año generalmente, de una región país o ciudad.

En la última década el sistema financiero boliviano ha tenido grandes cambios, pero uno de los mas importantes fue el diferimiento de la ley 1488 de 1993 y en su remplazo la ley 393 de 2013, en la cual se hace una nueva estructuración del sistema financiero.

Parece algo obvio e intuitivo que el crédito afecta de forma directa al crecimiento económico, por ejemplo, a mayor otorgamiento de crédito, mayor inversión y mayor crecimiento, pero existe una escasez de trabajos empíricos respecto a estos dos temas, en el caso boliviano, específicamente. El presente documento tratará de dar luz empírica a la forma en que se relaciona y el efecto que tendría un shock de crédito al crecimiento económico.

# 2. Revisión literaria.

Existe dos frentes teóricos respecto a la importancia del sistema financiero al crecimiento económico. Entre los más destacados, Schumpeter (1911) argumenta que el sistema financiero estimula la innovación y crecimiento, mediante la identificación y financiación a actividades productivas y por el otro lado Lucas (1988) que considera que el sistema financiero está sobrevalorado.

Existe una variedad de trabajos que tratan de dar luces sobre la relación entre el sistema financiero y el crecimiento económico, como el de King y Levine (1993) donde estudian datos de 80 países, investigando si existen correlaciones significativas y solidas entre el desarrollo financiero y las  tasas actuales y futuras de crecimiento, dicho documento llega a concluir que ‘… el porcentaje de crédito asignado a empresas privadas y la proporción de crédito emitidos a empresas privadas con respecto al PIB: están estrechamente relacionados con el crecimiento, la tasa de acumulación de capital físico y las mejoras en la eficiencia de la asignación de capital … los componentes predeterminados o predecibles de estos indicadores de desarrollo financiero están significativamente relacionados con valores posteriores de los indicadores de crecimiento.’ pag. 17.

De la misma manera de la cruz y Alcántara (2011) examinan la causalidad entre el crecimiento económico y el crédito bancario para México, una de sus conclusiones fue que el crédito sectorial que impacta a la economía es el de consumo y los servicios, no obstante, no rechazan la hipótesis de que el crédito total otorgado por la banca comercial causa un aumento en la actividad económica de México, bajo la misma línea Clavellina (2013) concluye en evidencia a favor de esta relación positiva, sin embargo, adiciona que esta no es una garantía y que dependen de otras variables. .

En el caso boliviano, el trabajo empírico de Morales (2007), Quiroz y Aguilar (2011) concluyen que el sistema financiero boliviano tiene un efecto positivo, pero modesto, sobre el crecimiento económico, Zenteno (2017) concluye bajo la misma línea, sin embargo, adiciona que un shock en la cartera bruta generaría una leve expansión al PIB.

Una vez ya desestacionalizadas las series, procederemos a hacer el analisis de estacionariedad, para verificar si las varibles tienen una media y covarianza constantes. A continuación se observa una tabla resumen de los test Dickey-Fuller Aumendata _ADF_ y la prueba Kwiatkowski – Phillips – Schmidt – Shin _KPSS_, con constante y con contante y tendencia.   



````python
def Stationarity_test (Serie, regre='c', criterio= None, test = 'adf'):
    """
    Devuelve los valores calculados de los test de raiz ADF y KPSS
    con asteriscos si el valor calculado del test es significativo:
    Para la prueba ADF:
        * 1%
        ** 5%
        *** 10%
    Para la prueba KPSS:
        * 1%
        ** 2.5%
        *** 5%
        **** 10%
    Parametros:
    Serie (Dataframe): Serie a estudiar 
    regre:(String): Se especifica si la regresion tiene:
                        'nc': no constante, no tendencia. 
                        'c': Solo constante (por defecto)
                        'ct: constante y tendencia'.
                        'ctt' tendencia constante y lineal y cuadrática.
                    Kpss solo acepta 'c' y 'ct'

    criterio (String): 'AIC', 'BIC', 't-stat' o None, para adf
                        'Legacy' o interger, para kpss
    test: (String): adf o kpss
    """
    
    from statsmodels.tsa.stattools import kpss
    from statsmodels.tsa.stattools import adfuller

    
    if test == 'adf':
        test = adfuller (x = Serie , regression = regre , autolag = criterio)
        v_calculado = test[0].round(2) 
        v_critico = list(test[4].values())

        j = 0 
        d = 0
        for i in v_critico:
            j += 1
            if v_calculado < i:
                d = i
                break
            else:
                continue
        if d == 0:
            cv = str (v_calculado)
        else:
            cv = '%s' % (v_calculado) + '*' * j
        
        return cv
    elif test == 'kpss':
        test = kpss (x = Serie , regression = regre, nlags = criterio)
        
        v_calculado = test[0].round(2) 
        v_critico = list(test[3].values())[::-1]
        
        j = 0 
        d = 0
        for i in v_critico:
            j += 1
            if  v_calculado < i:
                d = i
                break
            else:
                continue
        if d == 0:
            cv = str (v_calculado)
        else:
            cv = '%s' % (v_calculado) + '*' * j
        return cv
````

````python

def TablaEs (datos):
# Prueba de Raiz unitaria Dickey Fuller Aumentada con constate
    df = []
    for i in datos:
        ite =  Stationarity_test(Serie = datos[i], regre = 'c', test = 'adf', criterio = 'BIC')
        df.append(ite)

    adf_c = df

    # Prueba de Raiz unitaria Dickey Fuller Aumentada con constate y tendencia

    df = []
    for i in datos:
        ite =  Stationarity_test(Serie = datos[i], regre = 'ct', test = 'adf', criterio = 'BIC')
        df.append(ite)

    adf_ct = df

    # Prueba KPSS Aumentada con constate
    df = []
    for i in datos:
        ite =  Stationarity_test(Serie = datos[i], regre = 'c', test = 'kpss', criterio = 4)
        df.append(ite)

    kpss_c = df

    # Prueba de Raiz unitaria Dickey Fuller Aumentada con constate y tendencia
    df = []
    for i in datos:
        ite =  Stationarity_test(Serie = datos[i], regre = 'ct', test = 'kpss', criterio = 4)
        df.append(ite)

    kpss_ct = df


    # Crendo Tabla
    indice = list (datos.columns)
    pruebas = pd.DataFrame({'ADF_c': adf_c, 'ADF_ct': adf_ct,
                            'KPSS_c': kpss_c, 'KPSS_ct':kpss_ct}, index = indice)
    print('Pruebas de Estacionaridad')
    print(pruebas)
    print ('Para la prueba ADF: * 1%, ** 5%, *** 10%. \nPara la prueba KPSS: * 1% ** 2.5% *** 5%,**** 10%')
# criterio = 12
````

````python
TablaEs(datos = LDATA.drop(['LINV', 'LPIB'], axis = True))
````

Las otras variables no son estacionarias en ningun test, por lo tanto, se diferenciara la serie.

## 4.1. Modelo OLS

El siguiente cuadro hace referencia al modelo *1* especificado anteriormente, se puede evidenciar que la unica variable que tiene un impacto en
```SLPIB``` es el ```SLINV``` donde un crecimiento de 1% en la inversión provoca un crecimiento de 5% en el producto interno bruto, las demas variables son no significativas en el presente modelo.

El estadistico de bondad de ajuste es bajo, es decir las presentes variables explicativas solo explican el Crecimiento del PIB un 13%. Y según el estadistico Durbin-Watson el modelo no presenta autocorrelación de primer orden. Por otro lado, dado un nivel de significancia del 5% no rechazamos la hipotesis nula de no normalidad, es decir, nuestro modelo cumple el supuesto de normalidad, según el test Jarque-Bera.

````python
import statsmodels.formula.api as smf

# Funcion para generar lags de las series
def lag(x, n):
    if n == 0:
        return x
    if isinstance(x, pd.Series):
        return x.shift(n) 
    else:
        x = pd.Series(x)
        return x.shift(n) 

    x = x.copy()
    x[n:] = x[0:-n]
    x[:n] = np.nan
    return x

model01 = smf.ols(formula = 'SLPIB ~ LMORA + LCBRUTA + SLINV + LTASA + LPF + LM2',
                 data = LDATAD1).fit()
print(model01.summary())
````

#### Prueba Ljung-Box y Box-Pierce

La hipotesis nula de las pruebas LB y BP es no autocorrelación serial mientras que la hipotesis alterna si existe autocorrelación serial, como los p-value son mayores a 5% para el primer y segundo rezago, no existe autocorrelación.

````python
# Test LB Y BP
LB = sm.stats.acorr_ljungbox(model01.resid, return_df=True, boxpierce = True).iloc[:10]
print (np.round(LB, 2))
````

#### Prueba Breusch-Godfrey

Otra prueba formal es el estadistico de Breusch-Godfrey, el cual toma la misma hipotesis nula que las pruebas Ljung-Box y Box-Pierce.  
Dado que p-value es mayor que 5% se concluye que no existe correlación de primer orden. Esto refuerza el estadistico DW.

````python
from statsmodels.stats.diagnostic import acorr_breusch_godfrey as ABG
ResultadosTest = ABG(model01, nlags = 1)
Nombres = ["Estadístico LM", "p-value delestadísticoLM", "Estadístico F",
"p-value delestadísticoF"]
pd.Series(ResultadosTest, index = Nombres)
````

### Test Heterocedasticidad

#### Test Breush-Pagan

Una prueba ampliamente utilizada para detectar la heterocedasticidad es el estadistico  Breush-Pagan, el cual toma como hipotesis nula que el modelo es homocedastico y como hipotesis alterna heterocedasticidad. Como P-value es menor que alfa 5% se dice que existe presencia de heterocedasticidad, pero a 1% de significancia el modelo es homocedastico.

````python
from statsmodels.stats.api import het_breuschpagan as BP

# Test de Breusch - Pagan
    # H0 : Homocedasticidad  # H1 : Heterocedasticidad
Resultados_BP = BP(resid = model01.resid, exog_het = model01.model.exog)
names = ['Estadistico LM', 'p-value del estadistico', 'Estadistico F', 'p-value del estadistico F']
S1 = pd.Series (data= Resultados_BP, index=names)
R_BP = pd.DataFrame (data = {'Test de Breush-pagan': S1})
print (R_BP,'\nCon un alfa de 5%, dado p-value > alfa, no se rechaza H0, es homocedastica')
````

### Test multicolinealidad

La multicolinealidad se produce cuando dos o más variables exogenas tienen una relación exacta. A continuación se preseta el test factor de inflación de la varianza.   

#### Test factor de inflación de la varianza

````python
def FIV(modelo, inter = 1):
    """
    Realiza el test de factor de la inflacion de la varianza para todas las exogenas y
    retorna un DataFrame.
    Parametros:
    modelo (str): Indique el modelo donde guardo la regresión de STATSMODELS
    inter (int): 1 sin intercepto y 0 con intercepto, por defecto 1
    
    """
    from statsmodels.stats.outliers_influence import variance_inflation_factor as FIV # Importamos el modulo
    exog = pd.DataFrame(modelo.model.exog) # invocamos variables exogenas del modelo
    rango = exog.shape[1] # Rango columna de las exogenas
    Nombres = list(modelo.params.index) # Traemos los nombres de las exogenas
    vif = [FIV(exog = exog.values, exog_idx = i) for i in range (rango)] # Realizamos un bucle para analizar todo junto
    T_FIV = pd.DataFrame({'VIF': vif[inter:]}, index= Nombres[inter:]) # Crearmos un df
    return T_FIV # retornamos el df
````

Se observa que el modelo no tiene problemas de multicolinealidad

````python
print (FIV(model01))
print ('Si VIF > 10 existe problemas de multicolinealidad')
````
